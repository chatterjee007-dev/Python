## Write a program that allows the user to play a game of Rock Paper Scissors against the computer. The program should prompt the user to enter their choice (Rock, Paper, or Scissors) and then randomly generate a choice for the computer. The program should then determine the winner based on the following rules:

- ## Rock beats Scissors
- ## Scissors beats Paper
- ## Paper beats Rock
---
## **Importing module 'random' :**

- ### 'random' is a python module that generates random variables (random integers, bytes, sequences) based on the input that we give.
- ### In Rock Paper Scissors, as we randomly choose any one of the three, so we need to import random for that.
---
## **Defining our choices :**
- ### Creating a list of our choices.

```python
import random
choices = ['rock','paper','scissor']
```

### Now the logic to do this is : we have to enter our choice and the computer will randomly select one of these 3 choices > then we will define our rule, like - if 'scissor' > 'paper' > 'rock'.
## **Defining a function for computer's random choice :**
---
- ### **'random.choice'** - is a function present in module 'random' that randomly chooses one of the choices that we give from the list 'choices'.


```python
def get_computer_choice() :
  return random.choice(choices)
```

## **Defining a user function :**
---
- ### Asking user to enter a choice and using **lower()** function convert the input strings into lower vase as our list 'choices' are all in lower case.
- ### Now if the user input present in the list 'choices', then the function will return the user choice.

```python
def get_user_choice() :
  user_choice = input("Enter your choice, Rock paper or scissor? ").lower()
  if user_choice in choices :
    return user_choice
  else :
    return "Invalid choice. Please try again."
```

## **Applying game logic to determine the winner :**
---
- ### Defining function **'determine_winner( )'** below which is dependent on two parameters - user choice and computer choice.
- ### Now the variables '***user_choice***' and '***computer_choice***' are local variables which only apply for the function itself. So within the function we have to define what each of these variables stands for.
## **Let's define the logic within the function 'winner' :**
---
- ### If both user and computer choices are same - then it will be a tie.
- ### Then within **'elif'**  statement we will defines all the instances where the user can win.
- ### Otherwise the computer wins.

```python
def determine_winner(user_choice, computer_choice) :

    if user_choice == computer_choice :
        return "It's a tie!"

    elif (user_choice == 'rock' and computer_choice == 'scissor') or \
         (user_choice == 'scissor' and computer_choice == 'paper') or \
         (user_choice == 'paper' and computer_choice == 'rock') :
        return "You win!"

    else :
        return "Computer wins!"
```

## **Defining our game function - 'game_play' :**
---
- ### Now **' while loop'** keeps running until the user manually stops the program.
- ### Defining local variables **'user_choice'**  and  **'computer_value'**  which will call the two functions  - **'get_user_choice()'** and **'get_computer_choice()'** which we have defined earlier and the output will be assigned to them.

```python
def game_play() :
  print("Let's play!")

  while True :
    user_choice = get_user_choice()
    computer_value = get_computer_choice()
    print(f"You chose : {user_choice}")   # It will print the user and computer choice
    print(f"Computer chose : {computer_value}")
    result = determine_winner(user_choice, computer_value)  # Call function 'winner' to determine the winner
    print(result)

    play_again = input("do you want to play again? (y/n) : ").lower()
    if play_again != "y" :  # If the user don't want to play again, then the loop breaks
      break
```

## Let's run the function :

```python
game_play()
```

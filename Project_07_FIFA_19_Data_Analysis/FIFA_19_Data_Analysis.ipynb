## Let's see how our target variable ('Overall') is correlated to other variables using correlation matrix :

```python
# Calculating the correlation matrix of the numerical columns in the Pandas DataFrame 'data'.
correlation_matrix = data.corr(numeric_only=True)

# Calculating the correlation of the 'Overall' column with other columns and sorting the values in descending order.
correlation_overall = correlation_matrix['Overall'].sort_values(ascending = False)
print(correlation_overall)
```

## As we can see :
- #### **'Reactions'** (0.850071) and **'Composure'** (0.727646) have the highest positive correlations with Overall, suggesting that players with better reactions and composure tend to have higher overall ratings.
- #### **'Age'** has a moderate positive correlation (0.453354) which could imply that more experienced players generally have better overall ratings.
- #### Attributes like **'Balance'** have a very weak positive correlation, suggesting a minimal impact on the Overall rating.
- #### Goalkeeping attributes (**'GKPositioning'**, **'GKReflexes'**, etc.) have very weak negative correlations, which is expected as these attributes are specific to goalkeepers and don't directly influence the overall rating of outfield players.
- #### **'Jersey Number'** shows a weak negative correlation, indicating that it has little to no impact on a player's rating.
---
## visualizing the correlation matrix as a heatmap to show the relationships between different attributes in the FIFA 19 dataset :

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (20,20))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## The heatmap visualizes the correlation matrix, showing how different player attributes in the FIFA 19 dataset relate to each other.

- ### **Strong positive correlations (dark red) :** Attributes like **'Reactions'**, **'Composure**', **'Special**', and **'Vision**' have strong positive correlations with **'Overall**'. This suggests that players with higher values in these attributes tend to have higher overall ratings.
- ### **Moderate positive correlations (lighter red) :** **'ShortPassing**', **'BallControl**', and **'Dribbling**' have moderate positive correlations with **'Overall**'. These attributes are important for players, but their impact on the overall rating might be slightly less pronounced compared to attributes with stronger correlations.
- ### **Weak correlations (light colors) :** Attributes like **'Jersey Number**', **'Aggression**', and **'Interceptions**' show weak correlations with **'Overall**', suggesting they have minimal impact on the overall rating.
- ### **Negative correlations (blue) :** Goalkeeping attributes generally have negative correlations with Overall. This is expected as these attributes are specific to goalkeepers and don't directly influence the overall rating of outfield players.
- ### The heatmap helps identify which attributes are most important for determining the overall rating and understand the relationships between different attributes.

## Determining which attributes are most important for determining the overall rating :
- ### For this particular dataset we use **linear regression** to observe the most important attribute, because here we need to map out the importance of every feature and print the most important feature.
- ### Here our target variable is **'Overall'** (overall performance).

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Defining the features (x) and target (y) :
x = data.drop(['Overall','Name','Balance','Unnamed: 0'], axis = 1)
y = data['Overall']

# Converting string columns to numerical representations using one-hot encoding :
x = pd.get_dummies(x) # This will create new columns for each unique value in the string columns

# Defining and fitting the model :
model = LinearRegression()
model.fit(x,y)

# Creating a Pandas Series named 'feature_importance' that contains the coefficients of the linear regression
# model, sorted in descending order, to identify the importance of each feature in predicting the overall rating :
feature_importance = pd.Series(model.coef_, index = x.columns).sort_values(ascending = False)
print("Feature importance in determining overall rating : ")
print(feature_importance.head(5))
print(feature_importance.tail(5))
```

### As we can see : Player value is a major factor in determining overall rating. Higher values strongly correlate with higher ratings, suggesting a link between market worth, perceived potential, and in-game performance.
- ### As player value decreases, it has an increasingly negative impact on their overall rating. This suggests that lower-valued players tend to have lower ratings, likely due to a combination of less developed skills, lower potential, and limited market demand.

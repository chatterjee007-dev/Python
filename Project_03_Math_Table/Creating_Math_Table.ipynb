## Defining a function 'multiplication()' which will take an input 'num'.
- ### Here the logic is - we want to multiply the parameter 'num' by each number from 1 to 10.

```python
def multiplication(num) :
  for i in range(1,11) :  # Assigning numbers one by one to 'i' from 1 to 10 using for loop
    result = i*num
    print(i,'x',num,'=',result)  # multiplying each number from 1 to 10 with the input number and print the result

num = int(input("Enter a number : ")) # Taking input from user
multiplication(num)  # Calling the function we have defined
```

### So in the output we get multiplication table of 49 all the way upto 490.

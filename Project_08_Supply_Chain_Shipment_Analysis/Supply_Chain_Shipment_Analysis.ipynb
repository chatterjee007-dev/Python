```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset :
ship_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data_Science/Projects/Project Lists & Datasets/SCMS_Delivery_History_Dataset.csv')
ship_data.head()
```

## Printing a concise summary of the DataFrame 'ship_data' :

```python
print(ship_data.info())
```

## Generating descriptive statistics of the DataFrame 'ship_data' :

```python
print(ship_data.describe())
```

## Checking for null values in each column :

```python
ship_data.isnull().sum()
```

---
## Creating a **box plot** of all the numerical columns in the **'ship_data'** DataFrame to identify anomalies :

```python
plt.figure(figsize = (20,20))
sns.boxplot(data = ship_data.select_dtypes(include = 'number'))
plt.title('Boxplot of Numerical Columns')
plt.show()
```

### So we can see that the maximum number of outliers are in **'Line Item Value'**.
---
## Creating summary report that provides insights into the pricing trends and patterns within the supply chain :

```python
summary_report = """
    Summary Report :
    - Basic information about the dataset : {}
    - Summary statistics : {}
    - Missing values : {}
""".format(ship_data.info(), ship_data.describe(), ship_data.isnull().sum())
print(summary_report)
```
